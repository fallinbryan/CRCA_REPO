```python
import pandas as pd   
from sklearn.cluster import DBSCAN
import datetime as dt
from matplotlib import pyplot as plt
import numpy as np
```

```python
alr_df = pd.read_csv('data/AlarmTrainSet.csv')

alr_df = alr_df.drop_duplicates(subset='EventId', keep="first")

alr_df.set_index('EventId', inplace=True)
alr_df['EventDt'] = pd.to_datetime(alr_df['EventDt'])
alr_df['ClearDt'] = pd.to_datetime(alr_df['ClearDt'])
```

```python
alr_df['EVENT_DT_epoch'] = (alr_df['EventDt'] - dt.datetime(1970,1,1)).dt.total_seconds()
alr_df['EVENT_CLR_epoch'] = (alr_df['ClearDt']- dt.datetime(1970,1,1)).dt.total_seconds()
alr_df['Duration'] = alr_df['EVENT_CLR_epoch'] - alr_df['EVENT_DT_epoch']

alr_df.sort_values('EVENT_DT_epoch', inplace=True)
```

```python
X = alr_df[['EVENT_DT_epoch', 'Duration']].values
```

```python
db = DBSCAN(eps=9, min_samples=2, ).fit(X) #an alarm 7 seconds away is considered a neighbor in the same cluster
```

```python
labels = db.labels_
```

```python
len(labels)
```

```python
len(set(labels)) - (1 if -1 in labels else 0)
```

```python
alr_df['ClusterLabel'] = labels
```

```python
alr_df.to_csv('data/cleaned_dbscan2.csv')
```
